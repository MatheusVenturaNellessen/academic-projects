```python
import numpy as np
from scipy import stats
```

```python
np.random.seed(42)

lista = np.random.randint(1, 101, size=10)

moda, contagem = stats.mode(lista, keepdims=False)

lista
```

```python
print(f"Média: {np.mean(lista)}")
print(f"Mediana: {np.median(lista)}")
print(f"Moda: {moda}")
print(f"Ocorrências: {contagem}")
print(f"Variância: {np.var(lista):.2f}")
print(f"Desvio padrão: {np.std(lista):.2f}")
print(f"Amplitude: {np.ptp(lista)}")
```
