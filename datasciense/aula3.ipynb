```python
import numpy as np
```

```python
dados = np.random.normal(loc=0, scale=1, size=1000)
```

```python
import matplotlib.pyplot as plt

plt.hist(dados, bins=30, edgecolor="black", color="g", density=True)
plt.show()
```

```python
from scipy.stats import norm

dados = np.random.normal(loc=0, scale=0.1, size=1000)

plt.hist(dados, bins=10, density=True, alpha=0.6, edgecolor="black")

x = np.linspace(min(dados), max(dados), 1000)

pdf = norm.pdf(x, np.mean(dados), np.std(dados))

plt.plot(x, pdf, 'r-', linewidth=2, label="Tendência")
plt.title("Histograma de uma Distribuição Normal")
plt.xlabel("Valores")
plt.ylabel("Quantidade")
plt.legend()
plt.show()
```
